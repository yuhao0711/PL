```python
import pandas as pd

data = pd.read_csv("anime.csv")
```

```python
#Q1.過濾評分人數低於10000的資料並重新存取

data = data[data['members'] >= 10000]
data.to_csv('data_filter.csv', index=False)

data_filter = pd.read_csv("data_filter.csv")
```

```python
data_filter
```

```python
#Q2.將資料排序並重新存取

data_filter = data_filter.sort_values(by='rating', ascending=False)
data_filter.to_csv('data_sorted.csv', index=False)

data_sort = pd.read_csv("data_sorted.csv")
```

```python
data_sort
```

```python
#Q3.擷取其中排名前十的資料

data_head = data_sort.head(10)
```

```python
data_head
```

```python
#Q4.將內容翻譯成中文

import re
from googletrans import Translator

translator = Translator()

# 建立Translator物件
translator = Translator()

# 定義翻譯函式
def translate_text(text):
    # 將text轉換成字串
    text = str(text)
    # 翻譯並返回結果
    return Translator().translate(text, dest='zh-tw').text

data_translate = data_head.applymap(translate_text)

data_translate.to_csv('data_translate.csv', index=False)

data_translate = pd.read_csv('data_translate.csv')
```

```python
data_translate
```

```python
#Q5.翻譯修正(將作品名稱替換成中文譯名、將無法翻譯的文字替換掉)

#讀取英文名稱
name_list = data_head["name"].to_list()
print(name_list)

# 定義新名稱的list
new_names = ['你的名字', '鋼之煉金術師：Fullmetal Alchemist', '銀魂第四季', '命運石之門', '銀魂第三季', '排球少年！！烏野高中 VS 白鳥沢學園高中', '獵人（2011）', '銀河英雄傳說', '銀魂第二季延長', '銀魂劇場版 完結篇 永遠的萬事屋']

# 使用新名稱覆蓋name欄位內容
data_translate.loc[:, 'name'] = new_names

#替換掉無法翻譯的字
data_translate['genre'] = data_translate['genre'].str.replace('shounen', '少年')
data_translate['genre'] = data_translate['genre'].str.replace('Shounen', '少年')
```

```python
data_translate
```

```python
#Q6.篩選出最受歡迎的類型前三名
from collections import Counter
import itertools

#以逗號作切分
genre_list = data_translate['genre'].str.split('，')

# 將元組列表轉換成計數器物件
counter = Counter(itertools.chain(*genre_list))

# 取出前三多的元素，只包含元素名称
top3 = [element[0] for element in counter.most_common(3)]
```

```python
top3
```

```python
#Q7.找出同時具有前三受歡迎類型的作品

# 將 top_three 轉換為一個集合
top3_set = set(top3)

# 過濾出包含 top_three 中所有元素的子列表
filtered_data = data_translate[data_translate['genre'].apply(lambda x: all(i in x for i in top3_set))]
name_list = filtered_data['name'].tolist()
```

```python
name_list
```

```python
#Q8.不具有前三受歡迎元素的作品

#使用了波浪號(~)來排除包含任何一個元素的情況
filtered_data2 = data_translate[~data_translate['genre'].apply(lambda x: any(i in x for i in top3_set))]
name_list2 = filtered_data2['name'].tolist()
```

```python
name_list2
```

```python
#Q9.篩選出最不受歡迎的類型前三名

# 取出前三少的元素，只包含元素名称
last3 = [element[0] for element in counter.most_common()[-3:]]
```

```python
last3
```

```python
#Q10.不具有不受歡迎元素的作品

#使用了波浪號(~)來排除包含任何一個元素的情況
filtered_data3 = data_translate[~data_translate['genre'].apply(lambda x: any(i in x for i in last3))]
name_list3 = filtered_data3['name'].tolist()
```
